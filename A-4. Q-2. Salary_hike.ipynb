*Using transformation to make data into narmally distributed data.

```python
#using log_transformation
df_log = np.log(df_1)

#using sqrt tranformation
df_sqrt = np.sqrt(df_1)

#using cube root transformation
df_cbrt = np.cbrt(df_1)
```

```python
#plotting original data and tranformed data sets
plt.figure(figsize=(20,22))
#original data
plt.subplot(4,1,1)
sns.distplot(df_1);
plt.title('Original data')
#log_transformed data
plt.subplot(4,1,2)
sns.distplot(df_log);
plt.title('Log_transformed data')
#sqrt_transformed data
plt.subplot(4,1,3)
sns.distplot(df_sqrt);
plt.title('Sqrt_transformed data')
#cbrt_transformed data
plt.subplot(4,1,4)
sns.distplot(df_cbrt);
plt.title('cbrt_transformed data')
```

*After tranforming the data looks more normally distributed than before ,resulting in more accurate result.

*As a result the transformed data sets are more like bell shaped than original data.

```python
#regression plot of transformed data
plt.figure(figsize=(12,15))
#original data
plt.subplot(4,1,1)
sns.regplot(x='years_exp',y='salary_hike',data=df_1);
#log-transformed
plt.subplot(4,1,2) 
sns.regplot(x='years_exp',y='salary_hike',data=df_log,color='g');

#sqrt-transformed
plt.subplot(4,1,3)
sns.regplot(x='years_exp',y='salary_hike',data=df_sqrt,color='b');

#cbrt-transformed
plt.subplot(4,1,4)
sns.regplot(x='years_exp',y='salary_hike',data=df_cbrt,color='r');
```

*There is very minute change in the regression plot due to transformation as the regression plot of original data actually forming good fit.

```python
#correlation between x & y
print('correlation of original data')
print(df_1.corr())
print('********************')

print('correlation of log tranformed data')
print(df_log.corr())
print('********************')

print('correlation of square root transformed data')
print(df_sqrt.corr())
print('********************')

print('correlation of cube root transformed data')
print(df_cbrt.corr())
print('********************')
```

```python
#import statsmodels.formula.api
import statsmodels.formula.api as smf

#model
model=smf.ols("salary_hike~years_exp",data=df_1).fit()
#coefficient parametrs
model.params
#t-value and p-value
print(model.tvalues,'\n', model.pvalues)
#R^2 value and R^2_adj value
print(model.rsquared,model.rsquared_adj)
```

```python
#import statsmodels.formula.api
import statsmodels.formula.api as smf

#model
model_log=smf.ols("salary_hike~years_exp",data=df_log).fit()
#coefficient parametrs
model_log.params
#t-value and p-value
print(model_log.tvalues,'\n', model_log.pvalues)
#R^2 value and R^2_adj value
print(model_log.rsquared,model_log.rsquared_adj)
```

```python
#import statsmodels.formula.api
import statsmodels.formula.api as smf

#model
model_sqrt=smf.ols("salary_hike~years_exp",data=df_sqrt).fit()
#coefficient parametrs
model_sqrt.params
#t-value and p-value
print(model_sqrt.tvalues,'\n', model_sqrt.pvalues)
#R^2 value and R^2_adj value
print(model_sqrt.rsquared,model_sqrt.rsquared_adj)
```

```python
#import statsmodels.formula.api
import statsmodels.formula.api as smf

#model
model_cbrt=smf.ols("salary_hike~years_exp",data=df_cbrt).fit()
#coefficient parametrs
model_cbrt.params
#t-value and p-value
print(model_cbrt.tvalues,'\n', model_cbrt.pvalues)
#R^2 value and R^2_adj value
print(model_cbrt.rsquared,model_cbrt.rsquared_adj)
```

*Using model based on Original data set as it has highest r^2 value.

```python
#define dependent variable
y=df_1['salary_hike']
#define independent variable
x=df_1['years_exp']
```

# prediction for new data points

```python
years_exp_1 = pd.Series([4.5,2.9,3.1])
data_pred=pd.DataFrame(years_exp_1,columns=['years_exp'])
```

```python
model.predict(data_pred)
```

```python
preds = x[5:8]
preds
```

```python
#predicting using model.
pred_val=model.predict(preds)
pred_val
```

```python
actual_val=y[5:8]
actual_val
```

```python
#predicting using other model an caculationg rmse(root mean square error).
```

```python
#importing library
from sklearn.metrics import mean_squared_error
```

```python
#calculating for model
mse = mean_squared_error(actual_val,pred_val)
rmse = np.sqrt(mse) 
rmse
```

```python
#calculating rmse value for all model
# for model_sqrt
pred_val1 = model_sqrt.predict(preds)

#calculating mse
mse_sqrt = mean_squared_error(actual_val,pred_val1)

#calculating rmse
rmse_sqrt = np.sqrt(mse_sqrt)
print('rmse_sqrt:',rmse_sqrt)
```

```python
# for model_cbrt
pred_val3 = model_cbrt.predict(preds)

#calculating mse
mse_cbrt = mean_squared_error(actual_val,pred_val3)

#calculating rmse
rmse_cbrt = np.sqrt(mse_cbrt)
print('rmse_cbrt:',rmse_cbrt)
```

```python
# for model_log
predict_val2 = model_log.predict(preds)

#calculating mse
mse_log = mean_squared_error(actual_val,predict_val2)

#calculating rmse
rmse_log = np.sqrt(mse_log)
print('rmse_log:',rmse_log)
```

```python
#making data frame of rmse value
rmse_data = pd.DataFrame({'model_log':rmse_log,
                          'model':rmse,
                          'model_sqrt':rmse_sqrt,
                          'model_cbrt':rmse_cbrt},
                         index=[1])
rmse_data
```

```python
#predicting data using model_sqrt
input_ = x[5:8]
input_
```

```python
actual_ = y[5:8]
actual_
```

```python
input_sqrt = np.sqrt(input_)
predicted_ = model_sqrt.predict(input_sqrt)
predicted_
```

```python
np.square(predicted_)
```

Since, model_sqrt has the lowest rmse score we will use this model to predict salary hike.
